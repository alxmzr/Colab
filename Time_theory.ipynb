```python
from pandas.tseries.offsets import Minute
import pandas as pd
from datetime import datetime
from google.colab import drive

drive.mount('/content/drive')

file_path = '/content/drive/My Drive/EURUSD30.csv'
# file_path = '/content/drive/My Drive/EURUSD1.csv'

column_names = ["date", "time", "open", "high", "low", "close", "volume"]

df = pd.read_csv(file_path, names=column_names)

df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'])

start_time = pd.to_datetime('8:00')
end_time = pd.to_datetime('12:00')
filtered_df = df[(df['datetime'].dt.time >= start_time.time()) & (df['datetime'].dt.time <= end_time.time())]

filtered_df
```
