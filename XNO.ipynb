```python
import requests
import time

# Initialize previous prices for each ticker
previous_prices = {
    "BAN/XNO": None,
    "XRO/XNO": None,
    "XDG/XNO": None,
    "ANA/XNO": None
}

while True:
    tickers = ["BAN/XNO", "XRO/XNO", "XDG/XNO", "ANA/XNO"]

    for ticker in tickers:
        url = f"https://data.nanswap.com/price-changes?ticker={ticker}"
        response = requests.get(url)
        data = response.json()
        change_1h = data["1h"]["change"]
        change_1h_rounded = round(change_1h * 100, 2)

        print(f"Change 1 minute {ticker}: {change_1h_rounded}%")

        if previous_prices[ticker] is not None and change_1h_rounded == previous_prices[ticker]:
            print(f"Price for {ticker} has not changed")
        elif change_1h_rounded <= -1 or change_1h_rounded >= 1:
            telegram_url = f"https://api.telegram.org/bot6813450847:AAHEuqYB9TIugOkYRHUB4dOMKRfkgssrmz8/sendMessage?chat_id=1563343518&text=Change 1 minute for {ticker}: {change_1h_rounded}%"
            requests.get(telegram_url)

        previous_prices[ticker] = change_1h_rounded

    time.sleep(60)  # Wait for 1 minute before running the loop again
```
